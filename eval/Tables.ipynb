```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline
import pandas as pd
import re
pd.options.display.max_columns = None
pd.options.display.max_rows = 150
pd.set_option('display.float_format', lambda x: '%.1f' % x)

%run -i 'parse_results.py'
results['order_running_time'] = results['order_running_time'] / 1000000
results['avg_contraction_running_time'] = results['avg_contraction_running_time'] / 1000
results['prepro_running_time'] = results['order_running_time'] + results['avg_contraction_running_time'] / 1000
results['avg_customization_running_time'] = results['avg_customization_running_time'] / 1000
```

## Overview Orders

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
results \
    .loc[lambda x: x['machine'] == 'compute5'] \
    .loc[lambda x: x['optimizations'].isin(['none', 'tables'])] \
    .groupby(['graph', 'partition', 'order']) \
    ['order_running_time', 'avg_contraction_running_time', 'num_shortcuts', 'avg_customization_running_time', 'avg_query_time'] \
    .mean()
```

## Overview Optimizations

```python
results \
    .loc[lambda x: x['machine'] == 'compute5'] \
    .loc[lambda x: x['partition'] == 'ifc'] \
    .groupby(['graph', 'order', 'optimizations']) \
    ['avg_customization_running_time', 'num_triangles', 'avg_query_time'] \
    .mean()
```

## Order table

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x)

table = results \
    .loc[lambda x: x['machine'] == 'compute5'] \
    .loc[lambda x: x['optimizations'] == 'none'] \
    .loc[lambda x: x['partition'] == 'ifc'] \
    .loc[lambda x: x['order'] != 'cuts_reorder'] \
    .groupby(['graph', 'order']) \
    ['num_shortcuts', 'prepro_running_time', 'avg_customization_running_time', 'avg_query_time'] \
    .mean()

table = table \
    .reindex(axis='index', level=0, labels=['chicago', 'london', 'stuttgart', 'europe', 'europe_u100']) \
    .reindex(axis='index', level=1, labels=['orig', 'order_expansion', 'order_on_expanded', 'cuts']) \
    .rename(index={'chicago': 'Chicago', 'london': 'London', 'stuttgart': 'Stuttgart', 'europe': 'Europe-R', 'europe_u100': 'Europe-100'}) \
    .rename(index={'orig': 'Orig.', 'order_expansion': 'Exp.', 'order_on_expanded': 'Edge', 'cuts': 'Cut'})

table['num_shortcuts'] = table['num_shortcuts'] / 1000.
table['num_shortcuts'] = table['num_shortcuts'].round().astype(int) #.map('{:.0f}'.format)
table['avg_customization_running_time'] = table['avg_customization_running_time'].round().astype(int)
table['avg_query_time'] = table['avg_query_time'].round().astype(int)
latex = table.to_latex()
latex = re.sub(re.compile('([0-9]{3}(?=[0-9]))'), '\\g<0>,\\\\', latex[::-1])[::-1]
print(latex)
table
```

## Opt table

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x)

table = results \
    .loc[lambda x: x['machine'] == 'compute5'] \
    .loc[lambda x: x['partition'] == 'ifc'] \
    .loc[lambda x: (x['order'] == 'cuts') | ((x['order'] == 'cuts_reorder') & (x['optimizations'] == 'directed_inf'))] \
    .groupby(['graph', 'order', 'optimizations']) \
    ['num_triangles', 'avg_customization_running_time', 'avg_query_time'] \
    .mean()

table = table \
    .reindex(axis='index', level=0, labels=['chicago', 'london', 'stuttgart', 'europe', 'europe_u100']) \
    .reindex(axis='index', level=2, labels=['none', 'undirected_inf', 'directed_inf']) \
    .rename(index={'chicago': 'Chicago', 'london': 'London', 'stuttgart': 'Stuttgart', 'europe': 'Europe-R', 'europe_u100': 'Europe-100'})

table['num_triangles'] = table['num_triangles'] / 1000000.
table['avg_customization_running_time'] = table['avg_customization_running_time'].round().astype(int)
table['avg_query_time'] = table['avg_query_time'].round().astype(int)

table.index = [table.index.get_level_values(0), table.index.map(lambda index: { ('cuts', 'none'): 'None', ('cuts', 'undirected_inf'): 'Infinity', ('cuts', 'directed_inf'): 'Directed', ('cuts_reorder', 'directed_inf'): 'Reorder' }[(index[1], index[2])])]

latex = table.to_latex()
latex = re.sub(re.compile('([0-9]{3}(?=[0-9]))'), '\\g<0>,\\\\', latex[::-1])[::-1]
print(latex)
table
```
